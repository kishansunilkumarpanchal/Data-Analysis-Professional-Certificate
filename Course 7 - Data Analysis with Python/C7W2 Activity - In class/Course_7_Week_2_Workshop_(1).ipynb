```python
# Standard deviation is the measure of how spread out numbers are. 
# Pandas is a library in Python that is used to calculate the standard deviation.

# In Maths, the mean is defined as the average of all the given values. 
# It means that the sum of all the given values divided by the total number of values given. 
# Whereas, the standard deviation is considered as the measure of the dispersion of the data values from the mean.

import pandas as pd
s=[2,3,5,6,7,90,4,1,2,5]
df=pd.DataFrame(s)
df.describe()
```

```python
import numpy as np
#np.linspace() has two required parameters, start and stop, which you can use to set the beginning and end of the range: 
#creates evenly spaces array
bins=np.linspace(min(df["acceleration"]), max(df["acceleration"]), 4)
print(bins)
groupnames=['low','medium','high']
```

```python
df['acccategories']= pd.cut(df['acceleration'],bins,labels=groupnames)
```

```python
df.head(60)
```

```python
# Lets plot a histogram t understand distribution

#not a bell curve

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["price"],bins,rwidth=0.9)

# set x/y labels and plot title
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("count")
plt.pyplot.title("price ")
```

```python
import matplotlib.pyplot as plt

plt.bar(["low","medium","high"],df["pricecategories"].value_counts())
plt.xlabel("price",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Price bins",fontsize=15)
plt.show()
```

```python
# The pandas documentation describes qcut as a “Quantile-based discretization function.”
# This basically means that qcut tries to divide up the underlying data into equal sized bins. 
# The function defines the bins using percentiles based on the distribution of the data, not the actual numeric edges of the bins.

df['accqcut']=pd.qcut(df['acceleration'], q=3)
```

```python
#Use standard deviation to remove outliers.
# Standard deviation tells about how the values in the dataset are spread. 
# They also tells how far the values in the dataset are from the arithmetic mean of the columns in the dataset.


#2,3,4
upperlimit=df['acceleration'].mean()+2*df.acceleration.std()

print(upperlimit)
lowerlimit=df['acceleration'].mean()-2*df.acceleration.std()
print(lowerlimit)
```

```python
df[df.acceleration>upperlimit]
```

```python
df[df.acceleration<lowerlimit]
```

```python
#ZSCORE ... HOW MANY STD DEVIATION YOU ARE AWAY FROM THE MEAN
df['ZSCORE']= (df.acceleration-df.acceleration.mean())/df['acceleration'].std()
```

```python
df
```

```python
df[df['ZSCORE']>2.5]
```

```python
df[df['ZSCORE']<-2]
```
